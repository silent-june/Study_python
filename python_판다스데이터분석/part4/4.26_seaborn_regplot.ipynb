```python
# -*- coding: utf-8 -*-

#  
import matplotlib.pyplot as plt 
import seaborn as sns

# titanic  
titanic = sns.load_dataset('titanic')

#   (5: darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

#   (figure 2   )
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

#   -  (fit_reg=True)
sns.regplot(x='age',
            y='fare',
            data=titanic,
            ax=ax1)

#   -  (fit_reg=False)
sns.regplot(x='age',
            y='fare',
            data=titanic,
            ax=ax2,
            fit_reg=False)

plt.show()
```
