```python
# -*- coding: utf-8 -*-

import pandas as pd 

df = pd.read_excel('./남북한발전전력량.xlsx')

df_ns = df.iloc[[0,5], 3:]
df_ns.index = ['South', 'North']
df_ns.columns = df_ns.columns.map(int)  # map은 내 기억에,,,, 불러와서 변환한 다음 다시 선언까지 해주는 함수
```

```python
# 행, 열 전치하여 막대 그래프 그리기
tdf_ns = df_ns.transpose()
print(tdf_ns.head())
tdf_ns.plot(kind='bar')
```
